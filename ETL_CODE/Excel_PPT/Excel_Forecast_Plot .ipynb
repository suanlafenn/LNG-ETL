```python
import numpy as np
import matplotlib.pyplot as plt
from pptx import Presentation
from pptx.util import Inches 
import seaborn as sns
import io
from pptx.util import Inches, Pt,Cm
from pptx.enum.text import PP_ALIGN
```

```python
#Images
penguins = sns.load_dataset("penguins")
ax  = sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
image_0 = io.BytesIO()
ax.savefig(image_0)
#logo
logo = 'logo.jpg'
prs = Presentation()
#First slide
lyt=prs.slide_layouts[0] 
slide=prs.slides.add_slide(lyt)
left = top = Inches(0)
pic = slide.shapes.add_picture(logo, left, top)
title = slide.shapes.title
title.text = "Import & Export Forecasting"
subtitle = slide.placeholders[1]
subtitle.text = "Image Set"
#Second slide
layout = prs.slide_layouts[6]
slide = prs.slides.add_slide(layout)
left = top = Inches(0)
pic = slide.shapes.add_picture(logo, left, top)
top = Inches(2)
left = Inches(2.5)
height = Inches(5)
pic = slide.shapes.add_picture(image_0, left, top, height=height)
#Text box
left = Cm(7)
top = Cm(2.5)
width = Cm(15)
height = Cm(1)
txBox = slide.shapes.add_textbox(left, top, width, height)
tf = txBox.text_frame
p = tf.add_paragraph()
p.text = "Image Description"
p.font.size = Pt(40)



#Third slide
prs.save('test.pptx')
```
